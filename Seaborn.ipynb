```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Getting buildin Seaborn Dataset for practice
sns.get_dataset_names()
```

```python
# Example of how to load the dataset
dimon_df = sns.load_dataset('diamonds')
dimon_df.head()
```

```python
# Disttribution Plot

sns.displot(dimon_df['price'], kde = True, bins = 13)
# here kde means Kernal Density Estimation which is used to draw the line in data.
# bins means how many bars we want to show in the plot.
```

```python
# joint Plot
sns.jointplot(x='carat', y='price', data=dimon_df, kind='hex')
```

```python
# kde ploting
sns.kdeplot(dimon_df['price'])
```

```python
# Pair Plot
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue='sex', palette='coolwarm')
```

```python
# bar plot
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc = 0)

# hue is added to sagrigate more information in the plot
# just like adding males and females in the plot in this exmaple...
```

```python
# violin plot
# it is a combination of box plot and kde plot

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
# Strip plot
plt.figure(figsize = (8, 5))

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)

# jitter is used to avoid overlapping of data points
# Dodge is used to segregate the data points
```

```python
# Swarm Plot
sns.violinplot(x='day', y='total_bill', data=tips_df)

sns.swarmplot(x='day', y='total_bill', data=tips_df, color='black')
```

```python

sns.swarmplot(x='day', y='total_bill', data=tips_df)
```
